```python
plt.imshow(img) # OpenCV BGR Whereas Matplotlib its RGB
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('puppy.png',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
img_resize = cv2.resize(img,(500,1024))
```

```python
plt.imshow(img_resize)
```

```python
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(img, (0,0), fx=w_ratio, fy=h_ratio)
```
